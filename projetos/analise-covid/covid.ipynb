```python
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv'

vax = pd.read_csv(url)
vax.head()

# Informações gerais sobre o DF:
vax.info()
```

```python
print("\n Resumo estatístico das colunas numéricas: ")
vax.describe()
```

```python
vax['date'] = pd.to_datetime(vax['date'], errors='coerce')
vax.set_index('date', inplace=True)
print("\n Data Frame após a conversão de date para index")
vax.head()
```

```python
paises = vax['location'].nunique()
print("\n Quantos países estão sendo representados nesse Data Frame?", paises)
```

```python
print("\n Informações gerais após a manipulação de datas")
vax.info()
```

```python
# Análises combinatórias
df_vacinacao = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv")
df_demografia = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")

df_vacinacao['date'] = pd.to_datetime(df_vacinacao['date'], errors='coerce')
df_vacinacao.set_index('date', inplace=True)
```

```python
vacinacao_mais_recente = df_vacinacao.groupby('location').last().reset_index()
vacinacao_mais_recente.head(10)
```

```python
df_demografia = df_demografia[['location', 'population']].drop_duplicates(subset=['location'], keep='first')
df_merged = pd.merge(vacinacao_mais_recente, df_demografia, on='location', how='left')
df_merged['porcentagem_vacinada'] = (df_merged['total_vaccinations'] / df_merged['population']) * 100
top_10_vacinados = df_merged.sort_values(by='porcentagem_vacinada', ascending=False).head(10)

print("\n Top 10 países com maior porcentagem de vacinas:")
top_10_vacinados[['location', 'porcentagem_vacinada']]
```
